# Import Modules

```python
import pyFish
import pyFish.metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

# Load Sample data

```python
data = np.loadtxt('data/sample_vel_x_vel_y.csv', delimiter=',')
vel_x = data[:,0]
vel_y = data[:,1]
tint = 0.12

X1 = np.sqrt(np.square(vel_x)+np.square(vel_y))
X1 = pyFish.metrics().interploate_missing(X1)
```

# Initialize object with parameters

```python
sde = pyFish.Characterize(inc=0.01, inc_x=0.1, inc_y=0.1,t_lag=1000, simple_method=True, max_order=10, delta_t=1)
```

# Analyse

```python
out = sde([vel_x, vel_y],t=None,dt='auto',t_int=tint)
out
```

```python
avgdriftX, avgdriftY, avgdiffX, avgdiffY, avgdiffXY, op_x, op_y = out.data()
len(avgdriftX), len(avgdriftY), len(avgdiffX), len(avgdiffY), len(avgdiffXY), len(op_x), len(op_y)
```
