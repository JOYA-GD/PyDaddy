```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import pyFish
import random
import numba
import tqdm
import time
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jan 17 12:25:39 2020

@author: Ashwin
"""

@numba.jit(fastmath=True)
def polarity(vector):
    Mx = np.sum(np.cos(vector))/len(vector)
    My = np.sum(np.sin(vector))/len(vector)
    return Mx, My, np.round(np.sqrt(Mx**2 + My**2), 3)

@numba.jit(nopython=True, fastmath=True)
def spontaneous_event(N, system_state, noise):
    l = np.random.randint(N)
    system_state[l] = system_state[l] + np.random.normal(-noise/2, noise/2)
    
@numba.jit(nopython=True, fastmath=True)
def get_order(t):
    time_frame = np.where(t == np.nanmin(t))
    return time_frame[0][0], time_frame
    
@numba.jit(fastmath=True)
def non_spontaneous_event(N, system_state, order):
    L = random.sample(range(0,N), order+1)
    ss = system_state[L]
    if order == 1:
        system_state[L[0]] = system_state[L[1]]
    else:
        groups = list(itertools.combinations(system_state[L], order))
        gP = [polarity(g) for g in groups]
        max_p = groups[np.where(gP == np.amax(gP))[0][0]]
        for s in range(len(ss)):
            if ss[s] not in max_p:
                break
        system_state[L[s]] = np.mean(max_p)

@numba.jit(fastmath=True)
def update_time(t, time_frame, rates):
    t_ = t - t[time_frame]
    t_[time_frame] = np.random.exponential((1/rates[time_frame[0][0]]))
    return t_

def generate_ga_data(N=15, noise=1, rates=[0.1,2], iter_=10**6, time_step=0.12):
    system_state = np.random.uniform(0,2*np.pi, size = N)
    clock = 0
    M = []
    Mx = []
    My = []
    time_steps = []
    t = np.array([np.random.exponential((1/r), N) if r else np.ones(N)*np.nan for r in rates])

    for _ in tqdm.tqdm(range(int(iter_))):
        order, time_frame = get_order(t)
        if not order:
            spontaneous_event(N, system_state, noise)
        else:
            non_spontaneous_event(N, system_state, order)
        clock = clock + t[time_frame]
        if clock[0] >= time_step:
            time_steps.append(clock[0])
            clock = 0
            mx, my, m = polarity(system_state)
            Mx.append(mx)
            #Mx = np.append(Mx, mx)
            My.append(my)
            #My = np.append(My, my)
        t = update_time(t, time_frame, rates)
    return np.array(time_steps), np.array(Mx), np.array(My)
```

```python
start = time.time()
t, Mx, My = generate_ga_data(N=15, noise=np.pi, rates=[0.1, 2], iter_=10e6, time_step=0)
end = time.time()
print(end-start)
sns.distplot(Mx)
```

```python
np.savetxt('pairwise_highresolution.txt', np.vstack((t,Mx,My)).T)
```
