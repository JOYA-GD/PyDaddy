```python
def r2_vs_order(op, avgDrift, avgDiff, max_order=10):
    r2_drift = []
    r2_diff = []
    coeff_drift = dict()
    for i in range(max_order):
        p_drift, _ = metrics._fit_poly(x=op, y=avgDrift, deg=i)
        p_diff, _ = metrics._fit_poly(x=op, y=avgDiff, deg=i)
        r2_drift.append(metrics._R2(data=avgDrift, op=op, poly=p_drift, k=i, adj=True))
        r2_diff.append(metrics._R2(data=avgDiff, op=op, poly=p_diff, k=i, adj=True))
        coeff_drift[i] = p_drift
    return r2_drift, r2_diff, coeff_drift

def r2_vs_order_multi_dt(Mx, My, t_int=0.12):
    c = pyFish.Characterize([Mx, My],t=None, t_int=t_int)    
    max_dt = c._out.autocorrelation_time
    r2_drift_multi_dt = []
    r2_diff_multi_dt = []
    coeffs = dict()
    N = 8
    for n in range(1, N+1):
        _,_, avgDiff, avgDrift, op = sde._drift_and_diffusion(Mx, t_int, dt= int((n/N)*max_dt), delta_t=1, inc=0.01)
        r2_drift, r2_diff, coeff = r2_vs_order(op, avgDrift, avgDiff)
        r2_drift_multi_dt.append(r2_drift)
        r2_diff_multi_dt.append(r2_diff)
        coeffs[int((n/N)*max_dt)] = coeff
    r2_drift_multi_dt.append([int((i/N)*max_dt) for i in range(1,N+1)])
    r2_diff_multi_dt.append([int((i/N)*max_dt) for i in range(1,N+1)])
    return r2_drift_multi_dt, r2_diff_multi_dt, coeffs
```

```python
def plot_r2_vs_order_multi_dt(Mx, My):
    r2_drift_m_dt, r2_diff_m_dt, coeff_drift = r2_vs_order_multi_dt(Mx, My)
    fig = plt.figure(dpi=300)
    for i in range(len(r2_drift_m_dt)-1):
        plt.plot(range(10), r2_drift_m_dt[i], label=r2_drift_m_dt[-1][i])
    plt.legend()
    plt.title('R2_adj_vs_order')
    plt.xlabel('order')
    plt.ylabel('R2_adj')
    return fig, coeff_drift
```

```python
def plot_order_vs_coeff(coeff_drift, dt, k=0, ret=False):
    x = list(coeff_drift[dt].keys())
    y = []
    for d in coeff_drift[dt].keys():
        y.append(coeff_drift[dt][d][k])
    y = np.abs(y)
    if ret:
        return x, y
    plt.figure(dpi=300)
    plt.title('Order VS coeff of highest order, dt = {}'.format(dt))
    plt.ylabel('abs coeff')
    plt.xlabel('order')
    plt.plot(x,y)
    
def plot_order_vs_coeff_mult_dt(coeff_drift, k=0):
    dt = coeff_drift.keys()
    fig = plt.figure(dpi=300)
    for i in dt:
        x, y = plot_order_vs_coeff(coeff_drift, i, k=k, ret=True)
        plt.plot(x, y, label="dt={}".format(i))
    plt.legend()
    plt.title('Order VS coeff of {} order'.format(k))
    plt.ylabel('abs coeff')
    plt.xlabel('poly degree')
    return fig
```

```python
def plot_dt_vs_coeff(coeff_drift, deg, k = 0, ret=False):
    dt = list(coeff_drift.keys())
    y = []
    for i in dt:
        y.append(coeff_drift[i][deg][k])
    y = np.abs(y)
    if ret:
        return dt, y
    plt.figure(dpi=300)
    plt.plot(dt, y)
    plt.title('dt VS coeff, deg = {}'.format(deg))
    plt.xlabel('dt')
    plt.ylabel('abs coeff')

def plot_dt_vs_coeff_multi_order(coeff_drift, k=0):
    fig = plt.figure(dpi=300)
    dt_s = list(coeff_drift.keys())
    orders = list(coeff_drift[dt_s[0]].keys())
    for o in orders:
        x, y = plot_dt_vs_coeff(coeff_drift, deg=o, k = k, ret=True)
        plt.plot(x,y, label='order={}'.format(o))
    plt.legend()
    plt.title('dt VS coeff of {} order'.format(k))
    plt.xlabel('dt')
    plt.ylabel('abs coeff')
    return fig
```
