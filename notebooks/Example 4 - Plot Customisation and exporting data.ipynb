# Example 4: Plot customization and exporting data

```python
import sys
sys.path.append('..')
```

```python
import pyddsde
```

```python
# Load data
data, t = pyddsde.load_sample_dataset('model-data-vector-pairwise')
```

```python
ddsde = pyddsde.Characterize(data, t)
```

```python
ddsde.parameters()
```

```python
# Some plot customization, like changeing resolution, `dpi`, changing text font size and the number of y axis
# ticks can be done by passing the values to relevent parameters
timeseries_fig = ddsde.timeseries(start=1200,# stating index of the timeseries 
                                  end=3200, # end index of timeseries
                                  n_ticks=3, # number of y ticks
                                  tick_size=14, # both (x and y) axis tick label font size
                                  label_size=16, # axis label font size
                                  title_size=12, # title font size
                                  dpi=150 # figure resolution
                                 )
```

```python
# if `kde` is True, kernal density estimation will be fit to the histograms
histogram_fig = ddsde.histogram(kde=True)
```

```python
# Exporting data

# export data as csv and mat files, to disk.
# exported data includes, drift and diffusion calculated for all timescales inluded in slider range.
# Specify the valid `dir_path`, path to export data.
# if `dir_path` is None or not specified, a new folder named `results` will be created in current 
# working directory and data will be exported in here.
# `zip` will create an aditional zip file of the exported data.
# If `save_mat` is True, a mat of data, which can be loaded on MATLAB, will be included.

ddsde.export_data(dir_path=None, zip=True, save_mat=True)
```
